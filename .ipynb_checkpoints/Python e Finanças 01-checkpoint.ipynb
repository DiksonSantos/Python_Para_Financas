### Vamos analisar o IBOV

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader.data as web

# start= Ano-Mes-Dia
cotacao_ibov = web.DataReader('^BVSP', data_source='yahoo', start='2022-01-01', end='2022-05-05')# 85 dias de negociação.
display(cotacao_ibov)
```

### Analisando o Gráfico

```python
# Apenas a coluna Adj_Close 
cotacao_ibov['Adj Close'].plot(figsize=(15,10)) # Largura x Altura
```

### Retorno do IBOV

```python
# Percentual de variação -> Calcula Porcentagem de almento do Ultimo valor em relação ao primeiro.
# https://www.exponencialconcursos.com.br/macete-para-calcular-variacao-percentual
retorno_ibov = cotacao_ibov['Adj Close'][-1] / cotacao_ibov['Adj Close'][0]  - 1 

print(f'Retorno IBOV: {retorno_ibov:.2%}')
```

```python
lista = [1,2,3,4]
print(lista[-1])
```

### Analisando com Média Móvel

```python
cotacao_ibov['Adj Close'].plot(figsize=(15,10), label='IBOV__')

# Analisa blocos de 21 Itens:
cotacao_ibov['Adj Close'].rolling(21).mean().plot(label='Media Movel de 21 Dias:') # Media de 21 Dias


plt.legend() # Faz as Labels serem Mostradas no Gráfico
plt.show()
```
